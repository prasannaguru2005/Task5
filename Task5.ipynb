1.Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total = 0
    for i in likes:
        total += i
    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total)
    print("Post Status:", status)
analyze_engagement([300, 250, 200, 300])
```

3.Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0
    for r in rainfall_data:
        total += r
    average = total / len(rainfall_data)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)
analyze_rainfall([70, 75, 80, 63], 70)
```

4.Social Media – Duplicate Account Detection

```python
def check_duplicates(usernames):
    unique_users = set(usernames)
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
check_duplicates(["raj123", "anu45", "raj123", "kavi78"])
```

6.Agriculture – Premium Crop Price Filter

7.System Monitoring – Application Health Checker


```python
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)
check_system_health(7)    
```

9.E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
check_exam_eligibility(["P", "P", "A", "P", "P"])
```

10.Smart Electricity Bill Analyzer


```python
def electricity_bill(units):
    bill = 0

    # billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill < 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

total_bill, status = electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Level:", status)
```
